- For **binary or categorical data like ours, normality tests are not particularly meaningful. Instead, we focus on appropriate statistical tests for comparing proportions or differences between groups.** 
    </blockquote>

<a id="7.1 Chi-square Test for Proportions"></a>
# <left> <p style="background-color: #41627E; font-family:Hack; font-size:70%; color:white; text-align:center; border-radius: 0px; padding:10px;width:32%"> 7.1 Chi-square Test for Proportions</p> </left>

7.1.1 **Hypothesis :** 

<blockquote style="background-color: #C12283 ; padding: 18px; border-radius: 5px; color: white;">
   <li> H0: There is no significant difference in the distribution of 'VisitPageFlag' between the Control and Test groups.</li> 
   <li> H1: There is a significant difference in the distribution of 'VisitPageFlag' between the Control and Test groups.</li>  
 </blockquote>

#### 7.1.2 Chi-square Test :

```python
from scipy.stats import chi2_contingency

# Creating a contingency table
contingency_table = pd.crosstab(click_df['TestControl_Groups'], click_df['VisitPageFlag'])

print(contingency_table)

# Performing the chi-squared test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Displaying the test statistic and p-value
print(f"Chi-squared Test Statistic: {chi2}")
print(f"P-value: {p}")

# Checking for statistical significance
alpha = 0.05
if p < alpha:
    print("The distribution of 'VisitPageFlag' is significantly different between groups.")
else:
    print("There is no significant difference in the distribution of 'VisitPageFlag' between groups.")
```

   <blockquote style="background-color: #C12283 ; padding: 19px; border-radius: 5px; color: white;">
 <li>Chi-squared Test Statistic:
        The chi-squared test statistic is a measure of how much the observed counts deviate from what would be expected if there were no association between the groups and the 'VisitPageFlag'.</li>
        In this case, the chi-squared test statistic is 133.14.

 <li> P-value:
        The p-value associated with the chi-squared test statistic is very small **(8.44e-31), much less than the conventional significance level of 0.05.</li>
        A small p-value indicates strong evidence against the null hypothesis.
   </blockquote>

7.1.3 **Interpretation:**  

 <blockquote style="background-color: #C12283 ; padding: 19px; border-radius: 5px; color: white;">
 <li> With such a small p-value, we reject the null hypothesis. The null hypothesis for the chi-squared test is that there is no significant difference between the groups and the 'VisitPageFlag'.</li>
 <li> We,  therefore, conclude that there is a significant difference between the groups (Control and Test) and whether users visited the page or not.</li>
 </blockquote>
